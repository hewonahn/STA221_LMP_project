### join LMP and weather station 

```python
LMP_weather_station=pd.merge(LMP_data, weather_stations, left_on='Location', right_on='node_id', how='inner')
```

```python
LMP_weather_station['Time'] = pd.to_datetime(LMP_weather_station['Time'], utc=True).dt.tz_convert(None)

LMP_weather_station['Time'] = LMP_weather_station['Time'].dt.floor('H')

LMP_hourly_avg = LMP_weather_station.groupby(['Time', 'Location', 'weather station', 'node_id'])['LMP'].mean().reset_index()
LMP_hourly_avg
```

### join LMP_weather station and weather_data

```python
LMP_hourly_avg['Time'] = pd.to_datetime(LMP_hourly_avg['Time'])
weather_data['time'] = pd.to_datetime(weather_data['time'])

merged_data = pd.merge(
    LMP_hourly_avg, weather_data, 
    left_on=['Time', 'weather station'], 
    right_on=['time', 'station_id'], 
    how='inner'
)
merged_data
```

## join gas, storage, grid_mix, load join, and demand forcast


```python
gas_storage=pd.merge(
    gas_price, storage, 
    left_on=['Time'], 
    right_on=['Time'], 
    how='inner')
```

```python

gas_storage_grid=pd.merge(
    gas_storage, grid_mix, 
    left_on=['Time'],
    right_on=['Time'], 
    how='inner')
```

```python

gas_storage_grid_load=pd.merge(
    gas_storage_grid, load, 
    left_on=['Time'],
    right_on=['Time'], 
    how='inner')

gas_storage_grid_load
```

```python
gas_storage_grid_load['Time'] = pd.to_datetime(gas_storage_grid_load['Time'], utc=True).dt.tz_convert(None)
gas_storage_grid_load['Time'] = gas_storage_grid_load['Time'].dt.floor('H')
gas_storage_grid_load_hourly_avg = gas_storage_grid_load.groupby(['Time']).mean().reset_index()
```
